```python
!pip install diffusers
!pip install accelerate

from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained("SG161222/Realistic_Vision_V5.1_noVAE", torch_dtype = torch.float32)
pipe = pipe.to("cuda")
```

```python
prompt = "boxing day celebration"

image = pipe(prompt, negative_prompt = "", num_inference_steps = 20, guidance_scale = 5, width = 512, height = 512)
image.images[0]
```
